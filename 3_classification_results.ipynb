```python
%matplotlib inline

from sklearn.metrics import confusion_matrix
import seaborn as sns; sns.set()
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['figure.constrained_layout.use'] = True
```

```python
# get class map from the csv file
csv_dir = '/home/changhongw/datasets/SOL-0.9HQ-PMT/SOL-0.9HQ-PMT_meta.csv'
df = pd.read_csv(csv_dir)
classes = df[['modulation technique', 'label']
            ].value_counts().index.to_list()
class_map = dict([x for x in sorted(classes, key=lambda x: x[1])])
```

## S1+S2

```python
feature_spec = 's1s2'

tech_name = list(class_map.keys())
y_true = np.load('results/testresluts_truth_pred_' + feature_spec + '.npy')[0,:]
y_pred = np.load('results/testresluts_truth_pred_' + feature_spec + '.npy')[1,:]

# Build confusion matrix
confusion = confusion_matrix(y_true, y_pred)
norm_confusion = confusion.T / confusion.astype(np.float).sum(axis=1)
norm_confusion = norm_confusion.T # normalize confusion matrix

plt.figure(figsize=(6,5.5))
sns.heatmap(norm_confusion, cmap = "Blues", square=True, annot=True,
            xticklabels=tech_name, yticklabels=tech_name)
plt.xticks(np.arange(len(tech_name)), tech_name, rotation=45, fontsize=11.5)
plt.yticks(fontsize=11.5)
plt.ylabel('True label', fontsize=12)
plt.xlabel('Predicted label', fontsize=12)
rcParams['axes.titlepad'] = 15

plt.savefig('results/confusion_matrix_' + feature_spec + '.png')
# plt.savefig('results/confusion_matrix_' + feature_spec + '.pdf')
```
